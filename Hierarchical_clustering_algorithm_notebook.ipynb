```python
dr = df[['Popularity',
'Duration (ms)']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler
```

```python
X_scaled = scaler.fit_transform(dr)
```

```python
X_scaled
```

```python
#pca
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(X_scaled)
```

```python
pca_scaled
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c='red')
```

```python
#agglomerative clustering >> make dendrogram
import scipy.cluster.hierarchy as sc

plt.figure(figsize = (20, 7))
plt.title("dendogram")

sc.dendrogram(sc.linkage(pca_scaled, method = "ward"))
plt.xlabel("Th sample data")
plt.ylabel("Euclidean distance")
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric="euclidean", linkage = "ward",compute_full_tree='auto')
```

```python
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c = cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
silhouette_coeff = []
for k in range(2, 10):
    agglo = AgglomerativeClustering(n_clusters=k, metric = "euclidean", linkage = "ward")
    agglo.fit(X_scaled)
    score = silhouette_score(X_scaled, agglo.labels_)
    silhouette_coeff.append(score)
```
